```python
x_year = df['iyear'].unique()
y_year_counts = df['iyear'].value_counts(dropna = True).sort_index()
plt.figure(figsize=(20,8))
sns.barplot(x=x_year,y=y_year_counts,palette ='Paired')
plt.xlabel("year with attack",fontsize=20)
plt.ylabel("total attacks in each year",fontsize = 20)
plt.title("Attacks in different Year",fontsize = 30)
plt.show()
```
